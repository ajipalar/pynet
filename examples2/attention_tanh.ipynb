```python
def unit_vector(x):
    return x/np.linalg.norm(x)
def radians(a, b):
    a_u = unit_vector(a)
    b_u = unit_vector(b)
    return np.arccos(np.clip(np.dot(a_u, b_u), -1.0, 1.0))
```

```python
x = np.random.rand(2048, 256)
x2 = x[0:len(x)//2,]
x = x[len(x)//2:len(x),]
assert len(x)==len(x2)
y = []
z = []
f = lambda a, b: radians(a, b)
for i, v in enumerate(x):
    assert len(x[i])==len(x2[i])
    #t = np.dot(x[i], x2[i])/np.sqrt(len(x[i]))
    t = f(x[i], x2[i])
    y.append(t)
    t = f(x[i], x[i])
    z.append(t)
    #t = np.dot(x[i], x[i])/np.sqrt(len)

def degrees(x):
    return x*360/(np.pi*2)
y = np.array(y)
y = degrees(y)
n=8
plt.figure(figsize=(n,n))
plt.hist(y, bins=16)
plt.title("Angular distribution of 256-dimensional vector pairs")
plt.xlim(0, degrees(np.pi/2))
plt.xlabel("degrees")
plt.text(1.4, 103, f"N={len(x)}\nmed {np.median(y)}\nstd {np.std(y)}")
plt.show()
```
