```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = pd.read_excel("TableS5.xlsx")
```

```python
d[(d['LB']<= 1237) & (d['LB']>=1236)]
```

```python
plt.figure(figsize=(10,8))
plt.title("The Escherichia coli Protein Abundance and variation in LB growth medium")
plt.scatter(d.loc[:, 'LB.2'], np.log10(d.loc[:,'LB']))
plt.ylabel("Log10(Abundance) (copies / cell)")
plt.xlabel("% cv")
print("median Log10(abundance) : %s %s %s\nmedian %s cv : %s %s %s" %(np.round(np.median(np.log10(d.loc[:, 'LB'])), decimals=2),
                                                           '+/-',np.round(np.std(np.log10(d.loc[:, 'LB'])),decimals=2),
                                                           '%', np.round(np.median(d.loc[:, 'LB.2']), decimals=2), '+/-', 
                                                                np.round(np.std(d.loc[:, 'LB.2'])))
      
     )
```
